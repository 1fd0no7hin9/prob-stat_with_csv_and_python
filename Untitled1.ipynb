```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('airAccs.csv')

times = pd.DataFrame(df, columns=['times'])
dead = pd.DataFrame(df, columns=['Dead'])

mean = dead.mean()
mean = "%.2f"%mean.values
median = dead.median()
median = "%.2f"%median.values
mode = dead.mode()
mode = "%0.2f"%mode.values[0][0]
std = dead.std()
std = "%0.2f"%std.values
```

```python
table = pd.DataFrame(data = {'mean':[mean],
                             'median':[median],
                             'mode':[mode],
                             'std':[std]})
table
```

```python
dead.plot.box()

dead.plot()
plt.ylabel('some numbers')
plt.xlabel('times')
```

```python
plt.scatter(times,dead, data = dead)
plt.ylabel('some numbers')
plt.xlabel('times')
```
